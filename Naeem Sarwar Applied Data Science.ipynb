```python
# Step 3: Handle missing values
# For simplicity, let's fill missing values with the mean of each column
df['Price_in_thousands'].fillna(df['Price_in_thousands'].mean(), inplace=True)
df['Engine_size'].fillna(df['Engine_size'].mean(), inplace=True)
df['Horsepower'].fillna(df['Horsepower'].mean(), inplace=True)
df['Fuel_efficiency'].fillna(df['Fuel_efficiency'].mean(), inplace=True)

# Step 4: Verify that missing values have been filled
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
df = pd.read_csv('Cleaned_Car.csv')

# Set the style for the plots
sns.set(style="whitegrid")

# Plot 1: Scatter plot of Horsepower vs Price_in_thousands
plt.figure(figsize=(5, 4))
sns.scatterplot(x='Horsepower', y='Price_in_thousands', data=df, color='blue')
plt.title('Horsepower vs Price_in_thousands', fontsize=14)
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Price in Thousands ($)', fontsize=12)
plt.show()

# Plot 2: Bar plot for Fuel_efficiency by Engine_size
plt.figure(figsize=(5, 4))
sns.barplot(x='Engine_size', y='Fuel_efficiency', data=df, palette='viridis')
plt.title('Fuel Efficiency by Engine Size', fontsize=14)
plt.xlabel('Engine Size (L)', fontsize=12)
plt.ylabel('Fuel Efficiency (mpg)', fontsize=12)
plt.show()

# Plot 3: Histogram of Price_in_thousands
plt.figure(figsize=(5, 4))
sns.histplot(df['Price_in_thousands'], bins=15, kde=True, color='purple')
plt.title('Distribution of Price_in_thousands', fontsize=14)
plt.xlabel('Price in Thousands ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```
